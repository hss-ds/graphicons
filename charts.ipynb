# This is simple graphs we use day to day data analysis

We have simple implementations logic too.

Universal imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np

import squarify
```

### 1. Bar Chart

```python
months = ['January', 'Feburary', 'March']
sales = [20, 35, 30]

plt.bar(months, sales)

plt.xlabel("Month")
plt.ylabel("Sale")
plt.title("Monthly Sales Plot")

plt.show()
```

### 2. Line Chart

```python
rng = range(9,99)

x = sorted(random.sample(rng, 10))

y = [random.randint(1,2)*xi for xi in x]

plt.plot(x, y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary Chart')

plt.show()
```

### 3. Pie Chart

```python
Studies = ['History', 'Science', 'Maths']
students = [45, 30, 25]

plt.pie(students, labels=Studies)

plt.axis('equal')
plt.title('Student share per study')

plt.show()
```

### 4. Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')

plt.xlabel('Data')
plt.ylabel('Frequency')

plt.title('Title goes here')

plt.show()
```

### 5. Scatter Plot

```python
x = np.random.randn(15)
y = np.random.randn(15)

plt.scatter(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

### 6. Box Plot

```python
data = [np.random.randn(50)*100, np.random.randn(50)*100]

# Generating random data
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)

data = [data1, data2, data3]

fig = plt.figure(figsize =(10, 7))
 
# Creating box plot
plt.boxplot(data, patch_artist=True, labels=['Group 1', 'Group 2', 'Group 3'])

plt.title('Box Plot Example')
plt.ylabel('Value')
plt.show()
```

### 7. Heat Map

```python
data = np.random.randn(10, 20)
# print(data.size)
sns.heatmap(data)
plt.title('Heat Map')
plt.show()
```

### 8. Area Chart

```python
x = range(1,6)
y = [2,3,4,2,6]

plt.fill_between(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title Goes here')


plt.show()
```

### 9. Bubble Chart

```python
x = np.random.rand(25)
y = np.random.rand(25)
z = np.random.rand(25) * 1000  # Bubble sizes

plt.scatter(x, y, s=z)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bubble Chart Example')
plt.show()
```

### 10. Tree Map

```python
sizes = [25, 50, 75, 100]
labels = ['Quarter 1', 'Quarter 2', 'Quarter 3', 'Quarter 4']

squarify.plot(sizes, label=labels, alpha=0.7)
plt.axis('on')
plt.title('Tree Map Example')
plt.show()
```

### 11. Radar Chart

```python
labels=np.array(['Category 1', 'Category 2', 'Category 3', 'Category 4'])
stats=np.array([20, 34, 30, 35])

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()

stats=np.concatenate((stats,[stats[0]]))
angles+=angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='red', alpha=0.25)
ax.plot(angles, stats, color='red', linewidth=1)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Radar Chart Example')
plt.show()
```
