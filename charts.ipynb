# This is simple graphs we use day to day data analysis

We have simple implementations logic too.

Universal imports

```python
import matplotlib.pyplot as plt
import random
```

### 1. Bar Chart

```python
months = ['January', 'Feburary', 'March']
sales = [20, 35, 30]

plt.bar(months, sales)

plt.xlabel("Month")
plt.ylabel("Sale")
plt.title("Monthly Sales Plot")

plt.show()
```

### 2. Line Chart

```python
rng = range(9,99)

x = sorted(random.sample(rng, 10))

y = [random.randint(1,2)*xi for xi in x]

plt.plot(x, y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary Chart')

plt.show()
```

### 3. Pie Chart

```python
Studies = ['History', 'Science', 'Maths']
students = [50, 30, 20]

plt.pie(students, labels=Studies)

plt.axis('equal')
plt.title('Student share per study')

plt.show()
```
